- Code

#### always make a copy of pandas df before cleaning process

```python
patients_copy = patients.copy()
treatments_copy = treatments.copy()
adverse_copy = adverse.copy()
```

## a) Completeness Issues
- missing values of contact details of 12 patients in patients table
- only 280 rows are there instead of 350 in treatments table
- hba1c_change column has missing values in treatments table

```python
# we can't do anything for first two so we will handle the last one
```

### Define

```python
# Subtract hba1c_end from hba1c_start to get values of hba1c_change
```

### Code

```python
treatments_copy['hba1c_change'] = treatments_copy['hba1c_start'] - treatments_copy['hba1c_end']
```

### Test

```python
treatments_copy['hba1c_change'].isnull().sum()
```

no null values left in this column

## Define

```python
# 2. three variables stored in 2 columns auralin and novodra columns
# we will use the melt function on these columns
```

## Code

```python
pd.melt(treatments_copy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'], var_name='treatment', value_name='dose')
```

```python
treatments_copy = pd.melt(treatments_copy, id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'], var_name='treatment', value_name='dose')
```

```python
treatments_copy = treatments_copy[treatments_copy['dose'] != '-']
```

```python
treatments_copy
```

```python
treatments_copy['lower_dose'] = treatments_copy['dose'].str.split('-').apply(lambda x:x[0])
treatments_copy['higher_dose'] = treatments_copy['dose'].str.split('-').apply(lambda x:x[1])
```

```python
treatments_copy.drop(columns=['dose'], inplace=True)
```
